```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
import sklearn
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/Users/rohithr/Desktop/segmentation/Datasets/ElectricCarData_Norm.csv")
df.head()
df.info()
```

```python
plt.figure(figsize=(30,10))
sns.lineplot(x='Brand',y = 'TopSpeed',data = df)
plt.title("brand v/s top speed")
plt.show()
```

```python
plt.figure(figsize=(25,20))
df['Brand'] = df['Brand'].astype('category')
sns.scatterplot(x='Brand',y = 'Model',data = df)
plt.show()
```

```python
plt.figure(figsize=(20,15))
df['Brand'] = df['Brand'].astype('category')
df['Brand'] = df['Brand'].cat.codes
sns.barplot(x='Brand',y = 'Model',data = df,order =df['Brand'].astype('category'))
plt.show()
```
