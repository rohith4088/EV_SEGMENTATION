Yellowbrick is a Python library that extends the scikit-learn library for machine learning. It provides tools for visualizing the performance of machine learning models, understanding feature importances, and selecting the best hyperparameters. The Yellowbrick.cluster module in Yellowbrick is focused on visualizations related to clustering algorithms.
KElbowVisualizer is a class within Yellowbrick.cluster that is used for visualizing the "elbow" or "knee" point in the data to help you determine the optimal number of clusters (k) for clustering algorithms, such as K-Means. The elbow point is the point on the plot where the within-cluster sum of squares (inertia) starts to level off. This indicates that adding more clusters doesn't significantly reduce the sum of squares, suggesting that you've found a reasonable number of clusters for your data.

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
m = KMeans()
visualizer = KElbowVisualizer(m, k=(2, 10)).fit(X)
visualizer.show()
 
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)    
```

Making clusters based on the brands and the price 

```python

# Visualising the clusters
plt.figure(figsize=(7,7))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'yellow', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Clusters of Electric Car')
plt.xlabel('Brand')
plt.ylabel('PriceEuro')
plt.legend()
plt.show()
```

```python
#agglomarative clustering
plt.figure(figsize=(15,7))
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('TopSpeed')
plt.ylabel('Euclidean distances')
plt.show()
```
